### 3. 다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오 (정답을 각 요구사항별로 입력할 필요 없이 3번 문제에 대해 1개의 클래스 정의 코드를 제시하면 된다.)

```python
class Counter:
    def __init__(self,count,step=1):
        self.count=count
        self.step=step
        
    def __str__(self):
        return "[Count (step:" + str(self.step) + ")]" + str(self.count)
    
    def incr(self):
        self.count = self.count + self.step
    
    def __call__(self):
        return self.incr()
    
    def __add__(self,a):
        self.count += a
        return Counter(self.count,self.step)
    
    def __sub__(self,a):
        self.count -=a
        return Counter(self.count, self.step)
    
    def __cmp__(self,a):
        return self.count - a
    
    def __gt__(self,a):
        return self.count > a
    def __lt__(self, y):
        return self.count < y
    def __eq__(self,a):
        return self.count == a
    def __ne__(self,a):
        return self.count != a

print("요구사항 1")
c=Counter(10)
d=Counter(10,2)
print()

print('요구사항2')
print (c)
print(d)
print()

print('요구사항3')
c.incr()
d.incr()
print(c)
print(d)
print()

print('요구사항4')
c()
d()
print(c)
print(d)
print()

print('요구사항5')
c= c+5
d=d-5
print(c)
print(d)
print()

print('요구사항6')
print(c > 10)
print(d > 10)
print(c < 10)
print(d < 10)
print(c == 17)
print(d != 9)
print()
```

- __init__() 메소드를 통해 생성자에 count와 step 값을 초기화해준다.
- __str__() 메소드를 통해 기본 형식을 출력해준다. 이때, self.count와 self.step은 자료형이 int형이기 때문에 str을 사용하여 문자열로 바꾸어주어야한다.
- incr() 메소드를 통해 count에 step를 더해주어서 다시 count에 저장하게 도와준다.
- __call__() 메소드를 통해 return으로 incr() 메소드를 호출한다.
- __add__()와 __sub__() 메소드를 통해 더하기와 빼기 연산을 수치 연산자 중복으로 처리해준다.
- __gt__, __lt__,__eq__, __ne__를 통해 비교 연산자 중복을 해준다.

### 4. 다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.
- MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

그러면 for~in문 안에서 if e not in s:를 통해 인자로 받은 리스트에서 중복되는 것을 제외해서 새로운 리스트(s)를 만든 후, 이를 인자로 받은 리스트 l에 다시 집어넣는다. 마지막으로 str 메소드를 불러 형식을 갖춰 출력하게 한다. 즉 MySet클래스를 사용해서 리스트를 중복되는 것이 없는 집합으로 만들 수 있다.

```python
class MySet(list):
    def __init__(self,l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result
    
    @staticmethod
    def eliminate_duplicate(l):
        s=[]
        for e in l:
            if e not in s:
                s.append(e)
        l[:]=[]
        for e in s:
            l.append(e)
    
    def __or__(self,a):
        orList = MySet([])
        for x in self:
            orList.append(x)
        for x in a:
            orList.append(x)
        MySet.eliminate_duplicate(orList)
        return orList
    
    def __and__(self,a):
        andList = MySet([])
        for y in a:
            if y in self:
                andList.append(y)
        return andList
    
    def __sub__(self,a):
        subList = MySet([])
        for x in self:
            subList.append(x)
        for y in a :
            if y in subList:
                subList.remove(y)
        return subList
    
if __name__ == "__main__":
    s = MySet([1,2,2,3,6])
    print(s)
    t = MySet([2,3,4,5,6,7,8,8,8,8,8,9,9])
    print(t)
    print(t)
    w = s | t
    print(w)
    u = s&t
    print( u)
    m = s-t
    print(m)
```

교집합은 동일하게 들어있는 것들을 반환해주면 된다.

### 9. 다음과 같은 코딩이 가능하도록 SearchEngineWithOrderedWebWords안에 반복자와 관련된 메소드를 추가하시오.

```python
class SearchEngineWithOrderedWebWords(SearchEngine):
    def __init__(self,*args):
        SearchEngine.__init__(self,*args)
    def getWordsFrequency(self, reverse = False):
        M = []
        L = []
        m = {}
        p = []
        dM = {}
        Rev = False
        if reverse == False:
            Rev = True
        
        for x in self.urlList:
            M.append(htmlSiteDict(x))
        for x in M:
            for key in x.keys():
                for y in range(x[key]):
                    L.append(key)
        for k in L:
            m[k] = L.count(k)
        dM = sorted(m.items(),key = operator.itemgetter(1),reverse = Rev)
        return dM
    def __iter__(self):
        self.index = 0
        self.p = self.getWordsFrequency()
        return self
    def __next__(self):
        if self.index >= len(self.p):
            raise StopIteration
        n = self.p[self.index]
        self.index += 1
        return n
    
w4 = SearchEngineWithOrderedWebWords('http://www.times.com', 'https://github.com')
for i in w4:
    print(i)
```

iter() 메소드에서 index 와 getWordsFrequency() 메소드로 반환받은 튜플을 p에 저장한 후 self를 리턴하므로 next()메소드를 호출해 루프를 돌게된다. 
next()메소드에서는 index 값이 튜플의 길이보다 커질 때까지 하나씩 반환하므로 for in w4: print(i) 에서 튜플의 값이 하나씩 출력된다.
